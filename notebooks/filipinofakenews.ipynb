```python
from sklearn.model_selection import train_test_split

import pandas as pd
```

```python
fakenews_dataset = pd.read_csv('datasets/FakeNewsFilipino/FakeNewsFilipino.csv')
```

```python
fakenews_dataset
```

```python
from sklearn.utils import shuffle
fakenews_dataset.sample(frac=1, random_state=42)
```

```python
train_set, test_set = train_test_split(fakenews_dataset, train_size=0.7, random_state=42)
```

```python
train_set.to_csv('datasets/FakeNewsFilipino/train.csv', index=False)
test_set.to_csv('datasets/FakeNewsFilipino/test.csv', index=False)
```
