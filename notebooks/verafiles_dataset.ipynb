```python
import pandas as pd

dataset = pd.read_excel("/Users/jose/Documents/projects/machine_learning/MulticlassDataset/VF_complete_data.xlsx", "rated_with_quotes")
```

```python
dataset["RATING"].unique()
```

```python
filtered_dataset = dataset[dataset["RATING"].isin(["FALSE", "FAKE", "MISLEADING"])]
```

```python
cleaner_dataset = filtered_dataset.fillna("")
```

```python
cleaner_dataset["CONCAT QUOTES"] = cleaner_dataset.apply(lambda row: " ".join([str(row[f"QUOTE {i}"]) for i in range(1, 11)]), axis=1)
```

```python
final_dataset = cleaner_dataset[["RATING", "CONCAT QUOTES"]]
```

```python
final_dataset
```

```python
from sklearn.model_selection import train_test_split

# Splitting the dataset for each label
false_ds = final_dataset[final_dataset['RATING'] == 'FALSE']
false_ds_train, false_ds_test = train_test_split(false_ds, test_size=0.3, random_state=42)

fake_ds = final_dataset[final_dataset['RATING'] == 'FAKE']
fake_ds_train, fake_ds_test = train_test_split(fake_ds, test_size=0.3, random_state=42)

misleading_ds = final_dataset[final_dataset['RATING'] == 'MISLEADING']
misleading_ds_train, misleading_ds_test = train_test_split(misleading_ds, test_size=0.3, random_state=42)
```

```python
final_dataset
```

```python
false_ds_train, false_ds_val = train_test_split(false_ds_train[false_ds_train['RATING'] == 'FALSE'], test_size=0.25, random_state=42)
fake_ds_train, fake_ds_val = train_test_split(fake_ds_train[fake_ds_train['RATING'] == 'FAKE'], test_size=0.25, random_state=42)
misleading_ds_train, misleading_ds_val = train_test_split(misleading_ds_train[misleading_ds_train['RATING'] == 'MISLEADING'], test_size=0.25, random_state=42)
```

```python
train_dataset = pd.concat([false_ds_train, fake_ds_train, misleading_ds_train], ignore_index=True)
train_dataset
```

```python
val_dataset = pd.concat([false_ds_val, fake_ds_val, misleading_ds_val], ignore_index=True)
val_dataset
```

```python
test_dataset = pd.concat([false_ds_test, fake_ds_test, misleading_ds_test], ignore_index=True)
test_dataset
```

```python
train_dataset.to_csv("/Users/jose/Documents/projects/machine_learning/verafiles_dataset/train.csv", index_label=False)
test_dataset.to_csv("/Users/jose/Documents/projects/machine_learning/verafiles_dataset/test.csv", index_label=False)
val_dataset.to_csv("/Users/jose/Documents/projects/machine_learning/verafiles_dataset/val.csv", index_label=False)
```
